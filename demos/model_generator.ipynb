```python
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
pi = math.pi

def PointsInCircum(r, n=100):
    crange = np.linspace(0, 2 * pi, n, endpoint=False)
    return np.array([[math.cos(x) * r, math.sin(x) * r] for x in crange])

def TangentsToCircum(n=100):
    return np.linspace(360, 0, n, endpoint=False)

def PointsInSpiral(r, n, turns, ang=0):
    crange = np.linspace(0 + ang, 2 * turns * pi + ang, n)
    rads = np.linspace(0, r, n)
    return np.array([[math.cos(x) * rads[i], math.sin(x) * rads[i]] for i, x in enumerate(crange)])
```

```python
def print_model(arr, obj):
    for line in arr:
        print(f'Object : {obj}', end='')
        for num in line:
            print(f' {num:6.3f}', end='')
        print(';')
```

### Спираль вертикльная

```python
size = 50
turns = 1
radius = 0.87
scale = (1.0, 0.2, 1.0)

points = np.hstack((np.zeros((size, 1)), PointsInSpiral(radius, size, turns)))
points = np.vstack((points, np.hstack((np.zeros((size, 1)), PointsInSpiral(radius, size, turns, pi / 2)))))
points = np.vstack((points, np.hstack((np.zeros((size, 1)), PointsInSpiral(radius, size, turns, pi)))))
points = np.vstack((points, np.hstack((np.zeros((size, 1)), PointsInSpiral(radius, size, turns, pi * 3 / 2)))))
points *= scale

# intens = np.ones((size, 1))
intens = np.vstack((
    np.linspace(1.0, 2.0, size)[:,np.newaxis],
    np.linspace(1.5, 2.5, size)[:,np.newaxis],
    np.linspace(2.0, 3.0, size)[:,np.newaxis],
    np.linspace(2.5, 3.5, size)[:,np.newaxis]))

sizes = np.ones((size * 4, 1))
sizes = np.hstack((
    sizes * 0.05,
    sizes * 0.9,
    sizes * 0.05))

angles = np.zeros((size * 4, 1))

# Print
print(size * 4)
print_model(np.hstack((intens, points, sizes, angles)), 'cuboid')

plt.figure(figsize=(5,5))
plt.plot(points[:,1], points[:,2], 'ro')
plt.xlim((-1, 1))
plt.ylim((-1, 1))
```

### Волокна

```python
size = 400
side = 0.4
inter_rad = 0.3
limits = [[-side, side], [-side, side], [-0.6, 0.6]]

# Generate positions
pos = np.hstack((np.random.uniform(limits[0][0], limits[0][1], size=(size, 1)),
                 np.random.uniform(limits[1][0], limits[1][1], size=(size, 1)),
                 np.random.uniform(limits[2][0], limits[2][1], size=(size, 1))))

# Cut outlayers
pos = pos[np.sum(pos[:,:2] * pos[:,:2], axis=1) > inter_rad * inter_rad]

intens = np.ones(shape=(pos.shape[0], 1))
sizes_xy = np.random.uniform(0.01, 0.03, size=(pos.shape[0], 2))
sizes_z = np.random.uniform(0.1, 0.3, size=(pos.shape[0], 1))
sizes = np.hstack((sizes_xy, sizes_z))
angles = np.zeros(shape=(pos.shape[0], 1))


# Print
print(pos.shape[0])
print_model(np.hstack((intens, pos, sizes, angles)), 'paraboloid')

plt.figure(figsize=(15, 5))
plt.subplot(131)
plt.plot(pos[:,0], pos[:,1], 'ro')
plt.xlim((-1, 1))
plt.ylim((-1, 1))
plt.subplot(132)
plt.plot(pos[:,1], pos[:,2], 'ro')
plt.xlim((-1, 1))
plt.ylim((-1, 1))
plt.subplot(133)
plt.plot(pos[:,0], pos[:,2], 'ro')
plt.xlim((-1, 1))
plt.ylim((-1, 1))
```

```python
class CircularLayer:
    def __init__(self, radius, num, size, tangent_rotation_axis=-1, intens=1.0, rotation=[0.,0.,0.], height_offset=0):
        self.radius = radius
        self.num = num
        self.size = size
        self.tangent_rotation_axis = tangent_rotation_axis
        self.intens = intens
        self.rotation = rotation
        self.height_offset = height_offset

def GenerateCircularStructure(layers, is_vertical=False):
    components_num = 0
    for l in layers:
        components_num += l.num
        
        intens = np.zeros(shape=(l.num, 1)) + l.intens
        points = np.hstack([PointsInCircum(l.radius, l.num), np.zeros((l.num, 1)) + l.height_offset])
        sizes = np.repeat([l.size], l.num, axis=0)
        angles = np.repeat([l.rotation], l.num, axis=0)
        
        if l.tangent_rotation_axis != -1:
            angles[:, l.tangent_rotation_axis] += TangentsToCircum(l.num)
        
        if is_vertical:
            points[:,[0,1,2]] = points[:,[2,0,1]]
            angles[:,[0,1,2]] = angles[:,[2,1,0]]
            angles[:,[1,2]] += 90
            
        print_model(np.hstack((intens, points, sizes, angles)), 'ellipsoid')
    
    print(f'Components : {components_num};')
```

### Колесо с 10-ю спицами

```python
layers = [CircularLayer(0.9,  20, [0.17, 0.05, 0.05], 0), # Outer ring
          CircularLayer(0.5,  10, [0.19, 0.05, 0.05], 0), # Inner ring
          CircularLayer(0.7,  10, [0.20, 0.05, 0.05], 0, 1.0, [90.0, 0.0, 0.0]), # Outer spikes
          CircularLayer(0.38,  5, [0.10, 0.07, 0.07], 0, 1.0, [90.0, 0.0, 0.0]), # Inner spikes
          CircularLayer(0.21, 10, [0.05, 0.05, 0.07], -1, -1.0)] # Voids in disk
GenerateCircularStructure(layers, False)
```
